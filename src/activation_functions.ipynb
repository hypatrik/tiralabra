```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-5, 5.1, 0.1)
```

```python
# ReLu
from activation_funtions import ReLU

y = ReLU().activation(x)
plt.plot(x, y, label='relu-funktio')
plt.show()
```

```python
# Leaky ReLU
from activation_funtions import LeakyReLU

y = LeakyReLU(alpha=0.05).activation(x)
plt.plot(x, y, label='relu-funktio')
plt.show()
```

```python
# ReLu
from activation_funtions import Tanh

y = Tanh().activation(x)
plt.plot(x, y, label='tanh-funktio')
plt.show()
```

```python
# Sigmoid
from activation_funtions import Sigmoid

y = Sigmoid().activation(x)
plt.plot(x, y, label='sigmoid-funktio')
plt.show()
```
